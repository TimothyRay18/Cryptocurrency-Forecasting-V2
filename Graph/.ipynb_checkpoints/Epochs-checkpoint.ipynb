```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('../Model LSTM BTC/LSTM_BTC(32, 100, 50).pkl', "rb") as file_pi:
    lstm_btc = pickle.load(file_pi)
with open('../Model BiLSTM BTC/BiLSTM_BTC(32, 100, 50).pkl', "rb") as file_pi:
    bilstm_btc = pickle.load(file_pi)
with open('../Model LSTM ETH/LSTM_ETH(32, 100, 50).pkl', "rb") as file_pi:
    lstm_eth = pickle.load(file_pi)
with open('../Model BiLSTM ETH/BiLSTM_ETH(32, 100, 50).pkl', "rb") as file_pi:
    bilstm_eth = pickle.load(file_pi)
with open('../Model LSTM LTC/LSTM_LTC(32, 100, 50).pkl', "rb") as file_pi:
    lstm_ltc = pickle.load(file_pi)
with open('../Model BiLSTM LTC/BiLSTM_LTC(32, 100, 50).pkl', "rb") as file_pi:
    bilstm_ltc = pickle.load(file_pi)
    
fig, axs = plt.subplots(ncols=2, nrows=3)
fig.set_figwidth(12)
fig.set_figheight(12)
fig.tight_layout()

axs[0,0].plot(lstm_btc['loss'],label='train_loss')
axs[0,0].plot(lstm_btc['val_loss'],label='val_loss')
axs[0,0].set_xlabel('LSTM BTC')
axs[0,0].legend()
# axs[0,0].set_ylabel('MSE')
axs[1,0].plot(lstm_eth['loss'],label='train_loss')
axs[1,0].plot(lstm_eth['val_loss'],label='val_loss')
axs[1,0].set_xlabel('LSTM ETH')
axs[1,0].legend()
axs[2,0].plot(lstm_ltc['loss'],label='train_loss')
axs[2,0].plot(lstm_ltc['val_loss'],label='val_loss')
axs[2,0].set_xlabel('LSTM LTC')
axs[2,0].legend()
axs[0,1].plot(bilstm_btc['loss'],label='train_loss')
axs[0,1].plot(bilstm_btc['val_loss'],label='val_loss')
axs[0,1].set_xlabel('BiLSTM BTC')
axs[0,1].legend()
# axs[0,1].set_ylabel('MSE')
axs[1,1].plot(bilstm_eth['loss'],label='train_loss')
axs[1,1].plot(bilstm_eth['val_loss'],label='val_loss')
axs[1,1].set_xlabel('BiLSTM ETH')
axs[1,1].legend()
axs[2,1].plot(bilstm_ltc['loss'],label='train_loss')
axs[2,1].plot(bilstm_ltc['val_loss'],label='val_loss')
axs[2,1].set_xlabel('BiLSTM LTC')
axs[2,1].legend()
fig.subplots_adjust(wspace=0.15, hspace=0.15)
fig.text(0.5, -0.05, 'Epochs', ha='center', fontsize=12)
fig.text(-0.05, 0.5, 'Loss', va='center', rotation='vertical', fontsize=12)
# fig.legend(['loss', 'val_loss'], loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.08))
fig.savefig('../../Skripsi/img/lossVal_lossCompare.png', dpi=300, bbox_inches='tight')
```
