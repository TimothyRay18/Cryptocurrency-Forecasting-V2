```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

# load data
data = pd.read_csv('../Dataset/Binance_BTCUSDT_1h.csv')

# plot ACF
plot_pacf(data['Close'], lags=100)
plt.ylim(-0.1,0.1)
plt.show()
25
```
