```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

# load data
data = pd.read_csv('../Dataset/Binance_BTCUSDT_1h.csv')

fig, axs = plt.subplots()
# plot ACF
plot_pacf(data['Close'], lags=100)
plt.ylim(-0.1,0.1)
plt.savefig('../../Skripsi/img/PACF BTC.png', dpi=300, bbox_inches='tight')
```
