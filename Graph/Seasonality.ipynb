```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# load data
data = pd.read_csv('../Dataset/Binance_BTCUSDT_1h.csv')

# seasonal decomposition
result = sm.tsa.seasonal_decompose(data['Close'], period=960) 

# Get seasonal ACF
seasonal_acf = result.seasonal.diff().dropna().autocorr()

# Calculate lag
lag = int(round(1 / abs(seasonal_acf)))

print('Lag by seasonality:', lag)
plt.plot(result.seasonal)
plt.xlim(0, 960)
plt.show()
```
